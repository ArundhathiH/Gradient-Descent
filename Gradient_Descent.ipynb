```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grad_descent(x,y):
    m_current=0
    b_current=0
    iterations=500      #Tweak here
    l_rate=0.06        #Tweak here
    n=len(x)
    
    for i in range(iterations):
        y_pred = m_current*x + b_current
        cf = (1/n)*sum([j**2 for j in (y-y_pred)])  #refer cell below if doubt
        m_der = -(2/n)*sum(x*(y-y_pred))
        b_der = -(2/n)*sum(y-y_pred)
        m_current = m_current - l_rate*m_der
        b_current = b_current - l_rate*b_der
        print(f'm = {m_current}, b = {b_current}, CF = {cf}, iteration = {i+1}')
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

grad_descent(x, y)
```
